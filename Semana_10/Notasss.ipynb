## Numero de equipos ganadores de temporada

```python
df_Camp_season = df.sort_values(by="points", ascending = False).drop_duplicates(['season']).sort_values("season")[['season','club','points']]
df_Camp_season.index = df_Camp_season['season']
df_Camp_season.drop(columns='season', inplace = True)
df_Camp_season
```

## Equipos que mas veces ha ganado el tornedo

```python
#Resetea el index y muestra la columna de index anterior
df_camp = df_Camp_season.reset_index()
```

```python
#Agrupa segun una columna y se le agrega una funcion
df_camp[['club','season']].groupby('club').count().sort_values('season', ascending=False)
```

## Equipos que mas perdieron

```python
df_perdedores_por_temporada = df.sort_values('points').drop_duplicates(['season']).sort_values('season')[['season', 'club', 'points']]
df_perdedores_por_temporada[['club', 'season']].groupby('club').count().sort_values('season', ascending=False).iloc[0:5]
```

## 3 a√±os en el que se obtuvo mas puntaje

```python
df[['season','points']].groupby('season').sum().sort_values('points', ascending=False).iloc[:3]
```

## Equipos que mas han ganado en local

```python
df[['club','home_win']].groupby('club').sum().sort_values('home_win', ascending=False).iloc[:3]
```

## Equipos que han ganado mas encuentros por temporada

```python
df[['season','club','matches_won']].sort_values('matches_won', ascending=False).drop_duplicates(['season']).sort_values('season')
```

## Comparacion entre ganadores de local con away win

```python
df_analisis = df[['club','home_win','away_win']].groupby('club').mean()
df_analisis['home_vs_away'] = df_analisis['home_win'] / df_analisis['away_win']
df_analisis['home_status'] = df_analisis['home_vs_away'] > 1
df_analisis
```

## Graficos en pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.DataFrame(data=np.random.randn(365),
                 index = pd.date_range('2020/1/1', periods=365),
                 columns=['Stock Price'])
df = df.cumsum()
df.plot(figsize=(8,6))
plt.show()
```

```python
df = pd.DataFrame(data=np.random.randn(365,4),
                 index = pd.date_range('2020/1/1', periods=365),
                 columns=['Stock 1', 'Stock 2', 'Stock 3', 'Stock 4'])
df = df.cumsum()
df.plot(figsize=(8,6))
plt.show()
```

```python
df.plot(y='Stock 1', color='b', ylabel="Price [US$]", xlabel="Dates")
plt.show()
```

```python
df.plot(y=['Stock 1', 'Stock 2'], color=['b','g'], ylabel="Price [US$]", xlabel="Dates")
plt.show()
```

```python
df.plot(y=['Stock 1', 'Stock 2'], color=['b','g'], ylabel="Price [US$]"
        , xlabel="Dates", subplots=True,layout=(1,2), figsize=(12,6))
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(figsize=(12,4),nrows=1, ncols=2)

df.plot(y='Stock 1', ax=ax1)
df.plot(y='Stock 2', ax=ax2)
plt.show()
```
